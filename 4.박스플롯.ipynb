```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
```

```python
# 한글 설정
plt.rc('font', family='NanumGothic') # For Windows
print(plt.rcParams['font.family'])
```

```python
lst_name = ['수원','용인','부천','시흥','광명','안산','안양','포천','하남','화성']
```

```python
file_path = '클러스터링_데이터/클러스터링_용인.csv'
df = pd.read_csv(file_path)
```

```python
numeric_columns = ['입출차비율', '소매/유통', '의료/건강', '70세_이상_비율']
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(15, 10))

for ax, col in zip(axes.flatten(), numeric_columns) :
    sns.boxplot(x='kmeans_cluster', y=col, data = df, ax = ax)
    ax.set_title(f'클러스터별 {col} 박스플롯')
    ax.set_xlabel('클러스터')
    ax.set_ylabel(col)

plt.tight_layout()
plt.savefig('박스플롯/boxplot_용인.png')
plt.show()
```

```python
def plt_box_plot(name):
    file_path = f'클러스터링_데이터/클러스터링_{name}.csv'
    df = pd.read_csv(file_path)
    numeric_columns = ['입출차비율', '소매/유통', '의료/건강', '70세_이상_비율']
    fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(15, 10))

    for ax, col in zip(axes.flatten(), numeric_columns) :
        sns.boxplot(x='kmeans_cluster', y=col, data = df, ax = ax)
        ax.set_title(f'클러스터별 {col} 박스플롯')
        ax.set_xlabel('클러스터')
        ax.set_ylabel(col)

    plt.tight_layout()
    plt.title(f'{name} boxplot')
    plt.savefig(f'박스플롯/boxplot_{name}.png')
    plt.show()
```

```python
for name in lst_name:
    plt_box_plot(name)
```
