```python
#지역별 수정 필요 코드
df_care=pd.read_csv('/content/gdrive/MyDrive/BDA_안심데이터 공모전/반입_데이터/WellCareLongLat.csv')
df_care
```

```python
df_care.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df_care.head()
```

```python
df_pos = pd.DataFrame()
```

```python
df_pos
```

```python
df_pos = df_care.copy()
```

```python
df_pos = df_pos.rename(columns={'행정동코드':'행정동'})
```

```python
# 행정동코드에서 맨 뒤 두 자리를 제거
df_pos['행정동'] = df_pos['행정동'].astype(str).str[:-2]

# 결과 확인
print(df_pos)
```

```python
df_pos['행정동'] = df_pos['행정동'].astype(int) # 정수형으로 변환
```

```python
df_r = pd.merge(left = df,right = df_pos, how='inner',on='행정동')
df_r
```

```python
# 중복값 제거 후 인덱스 초기화
df_r['행정동'] = df_r['행정동'].drop_duplicates().reset_index(drop=True)

# 인덱스가 동일한 경우, 데이터가 부족할 수 있으므로, 데이터프레임의 길이에 맞추기
df_r = df_r.loc[df_r['행정동'].notnull()]
```
